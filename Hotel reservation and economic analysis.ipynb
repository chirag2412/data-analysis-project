```python
plt.figure(figsize=(10,6))
axl=sns.countplot(x='hotel',hue='is_canceled',data=df,palette='Blues')
legend_label,_=axl.get_legend_handles_labels()
#axl.legend(bbox_to_anchor(1,1))
plt.title('reservation status in diffrent hotel',size=20)
plt.xlabel('hotel')
plt.ylabel('no. of reservation')
plt.legend(['not cancled','cancled'])
plt.show()
```

```python
resort_hotel=df[df['hotel']=='Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize=True)
```

```python
city_hotel=df[df['hotel']=='City Hotel']
city_hotel['is_canceled'].value_counts(normalize=True)
```

```python
df.columns
```

```python
resort_hotel=resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel=city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize=(20,8))
plt.title('average daily rate in city and resort hotel',fontsize=30)
plt.plot(resort_hotel.index,resort_hotel['adr'],label='Resort Hotel')

plt.plot(city_hotel.index,city_hotel['adr'],label='City Hotel')

plt.legend(fontsize=20)
plt.show()
```

```python

plt.figure(figsize=(16,8))
axl=sns.countplot(x='arrival_date_year',hue='is_canceled',data=df,palette='bright')
legend_labels,_=axl.get_legend_handles_labels()
axl.legend(bbox_to_anchor=(1,1))
plt.title('reservation status per arrival_date_year',size=20)

plt.xlabel('arrival_date_year')
plt.ylabel('number of reservation')
plt.legend(['not cancled','cancled'])
plt.show()
```

```python
df['month']=df['reservation_status_date'].dt.month
plt.figure(figsize=(16,8))
axl=sns.countplot(x='month',hue='is_canceled',data=df,palette='bright')
legend_labels,_=axl.get_legend_handles_labels()
axl.legend(bbox_to_anchor=(1,1))
plt.title('reservation status per month',size=20)

plt.xlabel('month')
plt.ylabel('number of reservation')
plt.legend(['not cancled','cancled'])
plt.show()
```

```python

plt.figure(figsize=(16,8))
sns.barplot('month','adr',data=df[df['is_canceled']==1].groupby('month')[['adr']].sum().reset_index(),palette='bright')
plt.legend(fontsize=20)
plt.title('ADR per month',size=20)
plt.show()
```

```python
cancelled_data=df[df['is_canceled']==1]
top_10_country=cancelled_data[['country']].value_counts()[:10]
plt.figure(figsize=(16,8))
plt.pie(top_10_country,autopct='%.2f',labels=top_10_country.index)
plt.title('top 10 country with reservation cancle',size=20)
plt.show()
```

```python
cancelled_df_adr=cancelled_data.groupby('reservation_status_date')[['adr']].mean()
cancelled_df_adr.reset_index(inplace=True)
cancelled_df_adr.sort_values('reservation_status_date',inplace=True)
non_cancelled_data=df[df['is_canceled']==0]
non_cancelled_df_adr=non_cancelled_data.groupby('reservation_status_date')[['adr']].mean()
non_cancelled_df_adr.reset_index(inplace=True)
non_cancelled_df_adr.sort_values('reservation_status_date',inplace=True)
plt.figure(figsize=(16,8))

plt.title('average daily rate',size=20)
plt.plot(non_cancelled_df_adr['reservation_status_date'],non_cancelled_df_adr['adr'],label='non_cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'],cancelled_df_adr['adr'],label='cancelled')
plt.legend()
plt.show()
```
